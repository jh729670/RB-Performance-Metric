```python
import pandas as pd
import numpy as np
#Imports data
rb_raw=pd.read_csv('RB_Performance.csv')
```

```python
#Data Types
rb_raw.dtypes
```

```python
rb_raw.info()
```

```python
#Fills NA's with 0
rb_raw=rb_raw.fillna(0)

rb_raw.info()
```

```python
#COnverts Performance Metric to a float
rb_raw['Perfromance Metric']=pd.to_numeric(rb_raw['Perfromance Metric'],errors='coerce')

rb_raw.dtypes
```

```python
#Grabs features that I want to run machine learning on
data_rb=rb_raw.loc[:,['Player','GP','% of Snaps','# of Snaps','Total Touches','Total TD','100+ Yrd/games','Games w/TD','Top 12','Perfromance Metric']]

data_rb=data_rb.dropna()

data_rb.head()
#Target Variable
target=data_rb.iloc[:,-1]
#Feature Variable
features=data_rb.iloc[:,[0,1,2,3,4,5,6,7,8]]

col_names=data_rb.columns

features

```

```python
#Scales data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scale_data=scaler.fit(data_rb)
scale_data=scaler.transform(data_rb)


scale_data=pd.DataFrame(scale_data)
scale_data.columns=col_names
#Correlation Matrix 
corr_matrix=scale_data.corr()
#Correlation Matrix for Performance Metric
corr_matrix['Perfromance Metric'].sort_values(ascending=False)
```

```python
#Splits data into train and testing samples
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(features,target, test_size=.2, random_state=42)

player_test=X_test.loc[:,'Player']
player_train=X_train.loc[:,'Player']

X_train=X_train.drop(['Player'],axis=1)
X_test=X_test.drop(['Player'],axis=1)
```

```python
#Linear Regression Model
from sklearn.linear_model import LinearRegression
lm_model=LinearRegression()
lm_model.fit(X_train,Y_train)
lm_predict=lm_model.predict(X_test)

print('R^2 score :',lm_model.score(X_train,Y_train))
```

```python
#Cross Validation Score
from sklearn.model_selection import cross_val_score
cross_val_score(lm_model,X_train,Y_train,cv=10).mean()
```

```python
#Finds the mean squared error. Lower the number the less errors, hence more accurate
from sklearn.metrics import mean_squared_error

lm_mse=mean_squared_error(Y_test,lm_predict)

print('Mean Squared Error for Linear Regression :',lm_mse)
```

```python
#Create a dataframe of actual and predicted results


actual_predict=pd.DataFrame(list(zip(Y_test,lm_predict)),columns=['Actual Performance Metric','Predicted Performance Metric'])

actual_predict
```

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor

rf_model=RandomForestRegressor(n_estimators=100,oob_score=True,max_depth=3,random_state=42)

rf_model.fit(X_train,Y_train)

rf_predict=rf_model.predict(X_test)

```

```python
#Out of box score, similar to R^2 for linear regression
rf_model.oob_score_
```

```python
#Cross validation score for Random Forest Regressor
cross_val_score(rf_model,X_train,Y_train,cv=10).mean()
```

```python
#Mean Squared Error
rf_mse=mean_squared_error(Y_test,rf_predict)

print('Root Mean Squared Error Random Forest:',np.sqrt(rf_mse))
```

```python
#Dataframe of actual vs predicted results of Performance Metric
rf_actual_predict=pd.DataFrame(list(zip(Y_test,rf_predict)),columns=['Actual Performance Metric','Predicted Performance Metric'])

rf_actual_predict
```

I performed two machine learning techniques to determine how predictable Performance Metric is based of of several features. I used a multi-linear regression and a Random Forest Regression to compare which algorithm performed best. While both model performed similary, the Random Forest Regressor had a higher cross validation score and a lower mean squared error. Therefore, out of the two models, the Random Forest Regressor would be the perferred choice in predicted Performance Metric Accuracy.
